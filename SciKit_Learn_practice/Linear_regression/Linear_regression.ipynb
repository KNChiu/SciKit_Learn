```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import datasets 
```

```python
X,y = datasets.make_regression(n_samples=200,n_features=1,n_targets=1,noise=20)
```

```python
plt.scatter(X,y,linewidths=0.1)
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
predict = model.predict(X[:200,:])
```

```python
plt.plot(X,predict,c="red")
plt.scatter(X,y)
plt.show()
```

```python
from sklearn import metrics
import numpy as np

RMSE = np.sqrt(metrics.mean_squared_error(y, predict))
print("RMSE=", RMSE) # 使用RMSE查看
print("R2得分=", metrics.r2_score(y,predict)*100, '%') # 線性迴歸的衡量標準
```
